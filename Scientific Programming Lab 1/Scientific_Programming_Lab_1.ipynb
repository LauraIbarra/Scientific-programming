# 1. Context
<p align="justify"> According to the FAO, agriculture is the main source of water pollution due to the use of fertilizers, pesticides, and herbicides in cultivation systems. When these substances are not absorbed by plants or retained in the soil, they end up being washed away through agricultural runoff and drainage into surface and groundwater. This introduces compounds such as nitrogen, phosphorus, potassium, ammonium, organic matter, heavy metals including cadmium, lead, mercury, arsenic, chromium, and nickel, and pathogens, particularly Escherichia coli, Salmonella, helminths, viruses, and protozoa (FAO, 2021). Agriculture creates a cycle of contamination where the deteriorated water can be reused for irrigation, bringing with it a wide range of harmful substances that are not always measured and cannot be controlled. This not only affects crop health and results in poor-quality food production, but also poses a direct risk to the population when this contaminated water is consumed, exposing them to diseases such as diarrhea, cholera, hepatitis A, cancer, and polio (OMS, 2023).

<p align="justify"> To solve this problem, quality indicators are used. These are physical, chemical, and biological variables that help describe the state of the water and estimate the associated risk for its agricultural use and human consumption. Measuring these parameters in water can provide an estimate of the presence of contaminating nutrients, organic matter, salts, and sediments. For example, organic matter, which can be derived from agricultural waste and ends up in natural water sources, is determined by measuring Biochemical Oxygen Demand (BOD), Chemical Oxygen Demand (COD), and Total Organic Carbon (TOC). Nutrients contributed by fertilizers and pesticides are measured through nitrates (NO3−) and total nitrogen (TN). Turbidity measurement is related to sediment runoff, while electrical conductivity (EC) helps identify excess salts and ionic compounds in the water (Navarro et al., 2024; Abidin et al., 2024).

<p align="justify"> However, this measurement has its limitations as it doesn't precisely indicate which contaminant is present or its exact concentration, especially when it's at very low levels. For this reason, spectrometry is used to analyze the properties of a substance based on its interaction with light. In the case of water, contaminants can emit or absorb light, generating a spectral signature that allows them to be identified in the sample (Elsayed et al., 2020; Goblirsch et al., 2023). Integrating both physicochemical and spectral parameters allows not only for measuring the contaminant load in the water but also for precisely identifying which contaminants are present. This integration constitutes a monitoring tool that reduces the environmental and health risks associated with agricultural pollution.

# 3.	OOP for data processing

**First class**

<p align="justify"> To define the attributes of the SpectralProcessor class, "wavelengths" containing the wavelengths are utilized, and all rows containing the absorbance are extracted and stored in spectra.

```python
spectra = Absorbance[:, 1:]
print(spectra)
print(type(spectra))
```

<p align="justify"> The first method (apart from the constructor) will be a moving average smoothing. This is because the data are acquired with a UV-Vis spectrometer, and it is necessary to reduce the noise in the data, especially in the 200–300 nm region where peaks are generated due to the concentration of contaminants. This is done to obtain more stable and comparable curves without losing data from the peaks

<p align="justify"> The second method is normalization by area, with the purpose of making all spectra comparable to each other. For example, one sample may have a peak at 220 nm with a value of 0.8, while another has a value of 0.3. Although the spectral shape is the same, the difference in height can give the impression that they are distinct. With normalization, both curves are scaled to a total area of 1, and the comparison focuses on the shape of the spectrum (peaks and bands), not on the magnitude of the original values

<p align="justify"> The third method involves calculating the area under the curve. This approach was implemented because certain spectral ranges have a direct correlation with specific contaminants. For instance, the 200–250 nm range is associated with the concentration of NO3 and COD, while the 250–300 nm range is related to TOC. Integrating the area under the curve over these specific intervals summarizes the spectral information into a single, quantitative value. Consequently, this method facilitates facile comparison between samples and enables the correlation of these values with the presence of contaminants.

```python
class SpectralProcessor:

  def __init__(self, wavelengths, spectra):

    self.wavelengths = wavelengths
    self.spectra = spectra

  def smooth(self, window):

    kernel = np.ones(window) / window
    smoothed = np.array([np.convolve(row, kernel, mode="same") for row in self.spectra])
    self.spectra = smoothed
    return self

  def normalize(self):

     areas = np.trapz(self.spectra, x=self.wavelengths, axis=1)
     for i in range(len(areas)):
        a = areas[i]
        if a == 0:
            a = 1e-9
        self.spectra[i] = self.spectra[i] / a
     return self

  def bandArea(self,initial_nm, end_nm):

      mask_min = self.wavelengths >= initial_nm
      mask_max = self.wavelengths <= end_nm
      mask = mask_min & mask_max
      band_areas = np.trapz(self.spectra[:, mask], x=self.wavelengths[mask], axis=1)
      return band_areas
```

*Explanation of functions:*

<p align="justify"> convolve: This process, known as smoothing, is implemented using a moving average filter with NumPy. A kernel is constructed for this purpose and is then applied to each spectrum using the convolve() function. This replaces each value with the average of its neighbors within the specified window. The mode is set to 'same' to ensure the smoothed spectrum maintains the same number of data points, preventing any change in its dimensionality.

<p align="justify"> trapz: NumPy's trapz() function is utilized to compute the numerical integral of a dataset by employing the trapezoidal rule. In this specific application, the function is applied to each spectrum to estimate the area under the absorbance curve as a function of wavelength.

<p align="justify"> For the third method, np.trapz is also utilized to compute the area under the curve using the trapezoidal rule. This allows for the selection of a specific wavelength range (between initial_nm and end_nm), and it returns a single value representing the area within that band.

*To instantiate object of the first class:*

<p align="justify"> In this case, even though multiple methods will be used, a single object is instantiated to accumulate each of the data preprocessing steps. This approach ensures that all subsequent operations are performed on the same dataset, preserving the integrity and continuity of the entire process.

```python
spect_proc = SpectralProcessor(wavelengths, spectra)
```

**Second class**

<p align="justify"> The first method applied is the normalization of the physicochemical dataset. This is essential because the variables are expressed in diverse units and scales, which hinders direct comparison. For instance, parameters such as BOD, COD, turbidity, and electrical conductivity exhibit values of vastly different magnitudes. Without normalization, variables with wider numerical ranges could dominate the analysis and skew the interpretation. By normalizing the data, all variables are rescaled to a common reference, which allows for their equitable comparison and ensures that each parameter contributes proportionally to subsequent analyses.

<p align="justify"> The second method for outlier detection and correction was implemented to mitigate the effect of extreme outliers that distorted the distribution of the physicochemical variables. This is a frequent occurrence during in-field measurements. By replacing these values using the 1.5 * IQR rule, the dataset contains more robust and unbiased measurements, thereby preventing negative impacts on subsequent analyses and the interpretation of results.

```python
physicochemical = physicochemical_parameters.iloc[:, 1:].to_numpy(dtype=float)
print(type(physicochemical))

class physicochemicalProcessor:
  def __init__(self, physicochemical):

     self.physicochemical = physicochemical

  def normalizePhysicochemical(self, technique = "minmax"):

    try:
        if technique == "minmax":
          min = np.nanmin(self.physicochemical, axis = 0)
          max = np.nanmax(self.physicochemical, axis = 0)
          self.physicochemical = (self.physicochemical - min) / (max - min)
        elif technique == "zscore":
          mean = np.nanmean(self.physicochemical, axis = 0)
          std = np.nanstd(self.physicochemical, axis = 0, ddof = 1)
          self.physicochemical= (self.physicochemical - mean) / (std)
    except ValueError:
          print('unsupported technique')
    return self

  def removeOutliers(self):

     for j in range(self.physicochemical.shape[1]):
            col = self.physicochemical[:, j]
            q1 = np.percentile(col, 25)
            q3 = np.percentile(col, 75)
            iqr = q3 - q1
            lower = q1 - 1.5 * iqr
            upper = q3 + 1.5 * iqr
            mean_val =np.mean(col)

            self.physicochemical[:, j] = np.where((col < lower) | (col > upper), mean_val, col)
     return self
```

*Explanation of functions:*
<p align="justify"> For Min–Max normalization, nanmin and nanmax are used to identify the minimum and maximum values of each column. These values are then used to rescale each parameter to a range between 0 and 1. In the case of the Z-score technique, nanmean and nanstd are employed to calculate the mean and standard deviation of each variable, adjusting the values to be expressed in terms of their distance from the mean.

# 4.	Numerical derivation or integration of data.

<p align="justify"> In this case, a spectrometry database was used, from which the absorbance spectrum was differentiated with respect to the wavelength to more accurately identify the maximum and minimum points in the signal, namely the peaks and valleys of the spectrum. This technique allows for the highlighting of the most abrupt changes in the slope of the curve, which facilitates the precise location of absorption bands. Consequently, it enables the determination of the wavelength ranges where the contaminants present in the water samples are concentrated.

<p align="justify"> To represent the spectrum as a continuous function from discrete data, the function f_spectrum(x) was defined. This was achieved using np.interp, which linearly interpolates between the wavelength and absorbance values, allowing for the estimation of intermediate points. The values at the edges of the spectrum were fixed so that when the function is evaluated outside the measured range, it returns the value of the nearest endpoint instead of producing extrapolation errors.

```python
def f_spectrum(x):
    return np.interp(x, wavelengths, sample_four, left=sample_four[0], right=sample_four[-1])

print(sample_four[0])
print(sample_four[-1])
```

<p align="justify"> The classes Derivative and Der2central were implemented to calculate the slope of the absorbance spectrum. The former uses a forward-difference method, whereas the latter applies a central-difference method, which allows for a more precise estimation of the rate of change at each wavelength. This enabled the more accurate identification of the peaks and valleys present in the water sample's spectrum.

```python
class Derivative:
    def __init__(self, func, h=1e-5):
        self.f = func
        self.h = h
    def __call__(self, x):
        return (self.f(x + self.h) - self.f(x)) / self.h

class Der2central(Derivative):
    def __call__(self, x):
        return (self.f(x + self.h) - self.f(x - self.h)) / (2 * self.h)
```

<p align="justify"> The increment step, h, was defined as the average of the differences between the wavelengths in the database, using the diff function to obtain the differences and the mean function to calculate their average value. Subsequently, the Der2central class was instantiated with the spectral function f_spectrum and the step h. Finally, the derivative was evaluated at all points of the wavelengths vector (which are the absorbance values), resulting in the derivate_absorbance array. This array represents the rate of change of absorbance with respect to wavelength at each point of the spectrum.

```python
h = np.mean(np.diff(wavelengths))
df = Der2central(f_spectrum, h=h)
derivate_absorbance = df(wavelengths)
```

```python
print(f"The vector with the numerical values ​​derived from sample 4 are: {derivate_absorbance}")
plt.figure(figsize=(10,5))
plt.plot(wavelengths, derivate_absorbance, color="red" )
plt.xlabel("Wavelength (nm)")
plt.ylabel("derivate_absorbance (absorbance/nm)")
plt.title(f"Numerical derivative of the spectrum for sample 4")
plt.show()
```

<p align="justify"> The vector presented contains the numerical derivative values of absorbance with respect to wavelength for sample 4. Each value represents the slope of the spectrum: positive numbers indicate increasing absorbance, negative values reflect a decrease, and values near zero correspond to stable regions with no significant changes.

<p align="justify"> These variations are visualized in the graph. It is observed that the most abrupt changes occur in the region between 200 and 250 nm, which may correspond to the concentrations of NO3 and NT identified previously. As the wavelength increases, the derivative tends to stabilize near zero, indicating that the spectrum signal becomes more uniform and lacks marked variations. This suggests that no concentrations of contaminants are present in the water at these wavelengths.

# 5.	Matrix operations

<p align="justify"> In this step, the data structure was identified to ensure that both variables were defined as arrays, a transformation that had been performed previously from a DataFrame to NumPy. The wavelengths vector contains the spectral ranges, while spectra stores the absorbance values for all samples. By printing their data types and dimensions, it was confirmed that these variables were suitable for performing matrix operations. Additionally, the .shape attribute was used to show the size of the original absorbance matrix.

```python
print(type(wavelengths))
print(type(spectra))
print(f'The following vector shows the wavelengths are:\n {wavelengths}')
print(f'\nThe following matrix shows the absorbance values: {spectra}')
print(spectra.shape)
```

**slicing**: The chosen matrix operation is slicing, as it is very useful for extracting submatrices of interest from the original matrix.

```python
if spectra.shape[0] != wavelengths.shape[0]:
    matrix_absorbance = spectra.T
else:
    matrix_absorbance = spectra

mask = (wavelengths >= 200) & (wavelengths <= 300)
extract_wave_ranges= matrix_absorbance[mask, :]

print("The size of the extracted matrix in the range 200- 300 nm):", extract_wave_ranges.shape)
print("First 5 values of the extracted matrix:\n", extract_wave_ranges[:5, :5])
```

<p align="justify"> This operation consisted of extracting a specific spectral range. To do this, the dimensions of the wavelength vector and the spectra matrix were verified by comparing the number of rows of spectra with the length of wavelengths. If they did not match, the transpose "T" was applied so that each row corresponded to a wavelength and each column to a water sample. A boolean mask was then constructed to select the wavelengths between 200 and 300 nm, where contaminants such as NO3, NT, and COD were identified to be concentrated. The result is a reduced matrix that retains all samples but only includes the rows within the defined wavelength interval. The selection of this range is based on the graphical results from previous sections.

<p align="justify"> To avoid printing all 254 data points, only the first 5 rows of the resulting matrix were printed.

**<p align="justify"> centering**: The chosen matrix operation is centering, as it is very useful for adjusting the data by subtracting the mean of each column in order to standardize the scale of the matrix.

```python
extract_wave_ranges_centered = matrix_absorbance - matrix_absorbance.mean(axis=0, keepdims=True)

print("The size of the centered matrix: ", extract_wave_ranges_centered.shape)
print("First 5 values of the centered matrix:\n",extract_wave_ranges_centered[:5, :5])
```

<p align="justify"> For this second operation, matrix centering was applied. The mean function was used to calculate the average absorbance of each sample. The keepdims=True argument was specified to ensure that the resulting mean vector retained the same dimensional structure as the original matrix. By subtracting this mean vector from the original matrix, the centered matrix was obtained.

<p align="justify"> The result is a matrix with the same dimensions as the original (1546 rows × 29 columns), but its values are now adjusted relative to the mean of each sample. As shown in the table, the values are distributed around zero, which confirms that the centering was performed correctly. This is important because each sample's spectrum may have been shifted "upward" or "downward" since some samples absorbed more light on average than others. After centering, the comparison between samples is no longer based on which has the highest average absorbance, but instead focuses solely on how the shape of the spectrum varies across the wavelengths.

**<p align="justify">gram matrix:** The chosen matrix operation is the Gram matrix, as it is very useful for measuring the global similarity between all samples in the dataset.

```python
similar_spectra = extract_wave_ranges_centered.T @ extract_wave_ranges_centered

print("The size of the Gram matrix:", similar_spectra.shape)
print("First 5 values of the Gram matrix:\n", similar_spectra[:5, :5])
```

<p align="justify"> For this third operation, the Gram matrix was calculated. This was achieved by multiplying the centered matrix by its transpose using the matrix multiplication operator @. This operation produces a square matrix where each element represents the inner product between two samples, thereby measuring their global spectral similarity.

<p align="justify"> The resulting matrix has dimensions 29 × 29, corresponding to the number of samples. The values on the diagonal are large because they represent the comparison of each sample with itself, which always yields the maximum possible similarity. The off-diagonal values indicate the similarity between different samples. A high value signifies that the two samples have spectra with very similar shapes (i.e., their peaks and valleys coincide), whereas a low value indicates that the spectra have different shapes. This operation is significant because it condenses all pairwise similarities between samples into a single matrix.

**<p align="justify">eigen decomposition:** The chosen matrix operation is eigen decomposition, as it is very useful for identifying the main patterns of variation in the data through eigenvalues and eigenvecto

```python
eigvals, eigvecs = np.linalg.eig(similar_spectra)
lowest_highest= np.argsort(eigvals)[::-1]
eigvals = eigvals[lowest_highest].real
eigvecs = eigvecs[:, lowest_highest].real

print("The 5 main eigenvalues:\n", np.round(eigvals[:5], 4))
print("The 5 main eigenvectors:\n", np.round(eigvecs[:5], 4))
```

<p align="justify"> For this fourth operation, the eigenvalue and eigenvector decomposition of the Gram matrix was applied. The linalg.eig function was used to calculate the eigenvalues and eigenvectors of the square matrix. Subsequently, the argsort(eigvals)[::-1] method was employed to sort the eigenvalues in descending order, ensuring that the most significant patterns were prioritized. Using this same index, the eigenvectors were also reordered to maintain their correct association with their corresponding eigenvalues. Finally, the five largest eigenvalues and the first rows of their associated eigenvectors were printed.

<p align="justify"> The results show that the first eigenvalue, 583.36, is significantly larger than the others, which are 14.16, 6.23, 1.83, and 0.33. This indicates that the majority of the variation among the 29 sample spectra is concentrated in a single principal pattern, while the other components explain much smaller variations. The associated eigenvectors describe how each sample contributes to these patterns. In the resulting 29 × 29 matrix, each column corresponds to a component, and the values within it indicate the weight of each sample in that pattern. This makes it possible to identify which samples most closely follow the first pattern and which are related to the secondary ones. However, it's not possible to determine which specific contaminant corresponds to each pattern from the eigenvectors alone, although they likely relate to NT, NO3, or COD for each sample.

**<p align="justify">SVD:** The chosen matrix operation is Singular Value Decomposition (SVD), as it is highly useful for summarizing the information from numerous spectra into a few principal patterns. In this manner, instead of analyzing all wavelengths, it is possible to identify only the most important components that account for the majority of the differences among the samples.

```python
spectral_patterns, singular_values, sample_patterns = np.linalg.svd(extract_wave_ranges_centered, full_matrices=False)

print("The size of the spectral patterns:", spectral_patterns.shape)
print("The size of the singular values:", singular_values.shape)
print("The size of the sample  patterns:", sample_patterns.shape)
print("\nThe 5 main singular values:\n", np.round(singular_values[:5], 4))
```

<p align="justify"> For this fifth operation, Singular Value Decomposition (SVD) was applied to the centered matrix. This was achieved using linalg.svd function, which factors the matrix into three components: spectral patterns along the wavelengths, singular values that indicate the importance of each pattern, and sample-associated patterns. full_matrices=False is used to obtain the reduced version of the decomposition, avoiding the generation of matrices larger than the original.

<p align="justify"> The results show that the first singular values are significantly larger than the rest, which means that a few components summarize the majority of the variability among the spectra. This indicates that the principal differences among the 29 water samples are explained by a dominant pattern in the 200–300 nm region. As observed in previous analyses, this region is associated with contaminants such as NO3, NT, and COD, so the first SVD component likely reflects the presence and variation of these compounds in the samples. Although the full spectrum contains more than a thousand wavelengths, the essential information related to the contaminants is summarized in a very small number of principal patterns (200 - 300 nm).
